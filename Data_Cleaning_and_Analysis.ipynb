### Q.2 In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph.

### Q.10 What is the maximum duration of a Movie/Show on Netflix ?


```python

# Spliting data to get numeric value

df[['Duration_min', 'unit']] = df['duration'].str.split(' ', expand=True)

# Converting Object value of min from Object to Numeric

df['Duration_min'] = pd.to_numeric(df['Duration_min'])



# Maximimum Duration of a Movie

Movies = df[df['type'] == 'Movie']

Movies['Duration_min'].max()

# Maximum duration of a TV Show

TV_Show = df[df['type'] == 'TV Show']

TV_Show['Duration_min'].max()



```
